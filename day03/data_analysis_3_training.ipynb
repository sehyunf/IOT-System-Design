[연습문제 2]

연습문제 1에서 생성한 df_tst의 독립변수(temp_air, temp_gnd, humi, insol)와 종속변수(power)에 대한 선형회귀모델을 구축하고, 분석 결과를 출력해보시오.


```python
# 풀이
import statsmodels.api as sm

input_data = df_tst                                # 입력 데이터 설정
col_y = 'power'                                    # 종속변수 변수명
col_x = ['temp_air', 'temp_gnd', 'humi', 'insol']  # 독립변수 변수명
y = list(input_data[col_y])                        # 종속변수 데이터
x = input_data[col_x]                              # 독립변수 데이터

md = sm.OLS(y, x)                                  # 선형회귀모델 구축
md = md.fit()
print(md.summary())                                # 선형회귀분석 결과 출력
```

```python
# 풀이(1)
# 회귀계수 관련 정보 (변수)
var0 = 'const'
var1 = 'temp_air'
var2 = 'temp_gnd'
var3 = 'humi'
var4 = 'insol'
var = [var0, var1, var2, var3, var4]

# 회귀계수 관련 정보 (회귀계수, 위의 표에서 각 변수들의 coef 값)
B0 = 28.4338
B1 = -5.5996
B2 = 3.0129
B3 = 0.0337
B4 = 14.4172
B = [B0, B1, B2, B3, B4]

# 회귀계수 관련 정보 (P값, 위의 표에서 각 변수들의 P>|t| 값)
P0 = 0.265
P1 = 0.097
P2 = 0.207
P3 = 0.910
P4 = 0.111
P = [P0, P1, P2, P3, P4]

# 모형 적합도 관련 정보 (r2, P값)
r2_md = 0.653      # r2. 위의 표에서 Adj. R-squared
P_md = 2.73e-06    # P값, 위의 표에서 Prob (F-statistic)

print(f'표.다중선형회귀분석 결과')
print('-'*30)
print(f"{'변수':<9}{'회귀계수':<7}{'P값':<10}" )
print('-'*30)
for i in range(len(var)):
  print(f"{var[i]:<11}{B[i]:<11}{P[i]:<11.3f}" )
print('-'*30)
print(f"수정된 r2 = {r2_md}, P값 = {P_md:.3f}" )
print('-'*30)

print('')
print(f'* 구축한 다중선형회귀모델')
print('')
print(f'  power = {B[0]} + {B[1]}*{var[1]} + {B[2]}*{var[2]} + {B[3]}*{var[3]} + {B[4]}*{var[4]}')
```

```python
# 풀이(2) - 추가학습용
# 회귀계수 관련 정보 (변수)
var = list(md.params.index)

# 회귀계수 관련 정보 (회귀계수, 위의 표에서 각 변수들의 coef 값)
B = list(md.params.round(4))

# 회귀계수 관련 정보 (P값, 위의 표에서 각 변숮들의 P>|t| 값)
P = list(md.pvalues.round(3))

# 모형 적합도 관련 정보 (r2, P값)
r2_md = md.rsquared_adj.round(3)
P_md = md.f_pvalue.round(3)

print(f'표.다중선형회귀분석 결과')
print('-'*30)
print(f"{'변수':<9}{'회귀계수':<7}{'P값':<10}" )
print('-'*30)
for i in range(len(var)):
  print(f"{var[i]:<11}{B[i]:<11.4f}{P[i]:<11.3f}" )
print('-'*30)
print(f"수정된 r2 = {r2_md}, P값 = {P_md:.3f}" )
print('-'*30)

print('')
print(f'* 구축한 다중선형회귀모델')
print('')
print(f'  power = {B[0]} + {B[1]}*{var[1]} + {B[2]}*{var[2]} + {B[3]}*{var[3]} + {B[4]}*{var[4]}')
print('')

## 더 공부해 보면 좋을 것 (논문, 보고서 등에 선형회귀분석 사용 시 참고)
# 회귀계수 관련 : (1) VIF(Variance Inflation Factor), (2) 표준화된 회귀계수
# (1) VIF는 다중공산성 판단 기준으로, 10 보다 클 경우 해당 변수 제거하는게 좋음
# (2) 표준화된 회귀계수를 계산하여 제시할 경우에는 추가적인 분석도 가능함

# 선형회귀모델 관련 : (1) RMSE(Root Mean Square Error)
# (1) 구축한 선형회귀모델의 유용성을 나타내는 지표로 RMSE를 제시하기도 함
```
